```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Dispay the picture
plt.imshow(wizzy)
```

```python
# covert the picture from BGR to RGB 
#Display the changes
wiz = cv2.cvtColor(wizzy, cv2.COLOR_BGR2RGB)
plt.imshow(wiz)
```

```python
#resize the image
plt.imshow(cv2.resize(wiz,(300,300)))
```

```python
#resize the image to specific size
#display the height and width size
width = 0.5
height = 0.5
#give the new size a variable called wizr, then display what you want resized
wizr = cv2.resize(wiz, (0,0), wiz, width, height)
#display the new picture resized and the shape
plt.imshow(wizr)
print(wizr.shape)
```

```python
#flip the image -0 flips horizontally
img = cv2.flip(wiz,0)
plt.imshow(img)
```

```python
#flip the image -0 flips back to normal
img = cv2.flip(wiz, 3)
plt.imshow(img)
```

```python
#save image
cv2.imwrite(r'C:\Users\kemia\Desktop\wizzy.jpg',img)
```

# Drawing on Images

```python
#create a blank canvas
blank =np.zeros(shape =(600, 600,3), dtype = np.int16)
plt.imshow(blank)
```

```python
#Draw a rectangle on the blank canva -PT1(Top left Corner),PT2(Bottom right Corner)
cv2.rectangle(blank,pt1=(200,200), pt2=(400,400), color=(255,0,0), thickness=5)
plt.imshow(blank)
```

```python
#draw a green rectangle beside the red rectangle on the blank canva
cv2.rectangle(blank,pt1=(80,80), pt2=(180,500), color=(0,255,0), thickness=5)
plt.imshow(blank)
```

```python
#draw a rectangle on the blank canva
cv2.rectangle(blank,pt1=(410,405), pt2=(80,500), color=(0,255,0), thickness=5)
plt.imshow(blank)
```

```python
# place circle at the middle in blue colour not filled colour
cv2.circle(blank,center=(300,300), radius=50, color=(0,0,255), thickness=10)
plt.imshow(blank)
```

```python

# place circle at the middle in blue colour filled colour
cv2.circle(blank,center=(300,300), radius=50, color=(0,0,255), thickness=-1)
plt.imshow(blank)
```

```python
#create a blank canvas
blanka =np.zeros(shape =(600, 600,3), dtype = np.int16)
plt.imshow(blanka)
```

# Create a traffic light with the code in steps below

```python
# we create a white rectangle as we are trying to create a traffic light
cv2.rectangle(blanka,pt1=(150,50), pt2=(350,550), color=(255,255,255), thickness=5)
plt.imshow(blanka)
```

```python
# place circle at the top in green colour
cv2.circle(blanka,center=(250,150), radius=50, color=(0,255,0), thickness=-1)
plt.imshow(blanka)
```

```python
# place circle at the middle in yellow colour
cv2.circle(blanka,center=(250,300), radius=50, color=(255,255,3), thickness=-1)
plt.imshow(blanka)
```

```python
# place circle at the middlebottom below the middle in red colour
cv2.circle(blanka,center=(250,460), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blanka)
```
