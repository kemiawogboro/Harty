# Image processing using Numpy

```python

#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
#PIL -Python Imaging Library
from PIL import Image
```

```python
# we load the picture to read it into the python file note place the path by inputing location of picture file 
#and name with extention
messi=Image.open(r'C:\Users\kemia\Desktop\messilo.jpg')
```

```python
#we check what type of file is the picture saved as
type(messi)
```

```python
# we convert the file from a jpeg into a numpy array
# we call the array file
messia = np.asarray(messi)
messia
```

```python
# We check the Dimensionality (Height =1200, Width=992, Colour Channel=3 ,note we go by index so three colours willbe 3 even though the index is 0.1.2)
messia.shape
```

```python
#here we display the picture by placing the variable name of the array created in the plt.imshow as a parameter
plt.imshow(messia)
```

```python
# lets start with colour RED note the sequence of the colours in index order(Red=0, Green=1, Blue=2)
#code we want to display the concentration of red in the diagram 
plt.imshow(messia[:,:,0])
```

```python
#Grayscale
##let start with color green note the sequence of the colours in index order(Red=0, Green=1, Blue=2)
# we detect the green in grayscale
plt.imshow(messia[:,:,1], cmap= 'gray')
```

```python
# we want to display the red arrays only
messia[:,:,0]
```
