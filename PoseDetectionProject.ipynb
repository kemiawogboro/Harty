```python
#step 1 Import all necessary libraries
import cv2
import mediapipe as mp
```

```python
#step 2: i Identify your webcam
cap= cv2.VideoCapture(0) #cos I am using my local webcam
```

```python
#step 4 : leveraging the mediapipe library used for pose dertection

mpPose = mp.solutions.pose
pose = mpPose.Pose()
# pose =mpose.Pose(static_image =False, upper_body_only =False, smooth_Landmarks=True, min_detection_confidence=0.5, min_tracking_confidence=0.5)

# to draw and connect the landmarks
mpDraw = mp.solutions.drawing_utils
```

```python
#step3: Switch on webcam
while True:
    _, img = cap.read()
    
     #convert Image FROM BG to RGB 
    imgRGB  = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
     #Apply mediapipe
    results = pose.process(imgRGB)
    print(results.pose_landmarks)
    
    # Draw landmarks
    if results.pose_landmarks:
        
    
                mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
    
    cv2.putText(img, 'KemiAwosco Pose Detection Projects', (10, 40), cv2.FONT_HERSHEY_PLAIN, 3,(0,255,0), 2)
    cv2.imshow('KemiAwosco Pose Detection Projects', img)
    if cv2.waitKey(0) & 0xFF == ord('q'):# this allows us to kill the while loop
        break
        
#Release the capture once all the processing is done.
cap.release()
cv2.destroyAllWindows()
```
